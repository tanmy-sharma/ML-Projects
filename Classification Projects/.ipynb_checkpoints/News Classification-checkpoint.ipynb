## PreProcessing the data

```python
from string import punctuation

def html_del(text):
    soup = BeautifulSoup(text, "html.parser")
    soup = soup.get_text()
    return soup

def remove_brackets(text):
    return re.sub('\[[^]]*\]', '', text)

def remove_url(text):
    return re.sub(r'http\S+', '', text)

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    punctuations = list(string.punctuation)
    stop_words.update(punctuations)
    text = word_tokenize(text)
    filtered_words = [word for word in text if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
df['text'] = df['text'].apply(html_del)
df['text'] = df['text'].apply(remove_brackets)
df['text'] = df['text'].apply(remove_url)
df['text'] = df['text'].apply(remove_stopwords)
```

```python
tokenized_words = [word_tokenize(text) for text in df['text']]
corpus = [word for sublist in tokenized_words for word in sublist]
```

```python
punctuation_extra = ['’','”','“']
corpus = list(filter(lambda a: a not in punctuation_extra, corpus))
```

```python
counter = Counter(corpus)
most_common = counter.most_common(10)
most_common = dict(most_common)
most_common
```

```python
def nGramCalculator(corpus, n, g):
    vec = CountVectorizer(ngram_range = (g,g)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis = 0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x : x[1], reverse = True)
    return words_freq[:n]
```

**Unigram Analysis**

```python
plt.figure(figsize = (8,8))
most_common = dict(nGramCalculator(corpus, 10, 1))
most_common
sns.barplot(x = list(most_common.values()), y = list(most_common.keys()))
```

**Bigram Analysis**

```python
plt.figure(figsize = (8,8))
most_common = dict(nGramCalculator(corpus, 10, 2))
most_common
sns.barplot(x = list(most_common.values()), y = list(most_common.keys()))
```

**Trigram Analysis**

```python
plt.figure(figsize = (8,8))
most_common = dict(nGramCalculator(corpus, 10, 3))
most_common
sns.barplot(x = list(most_common.values()), y = list(most_common.keys()))
```

## Splitting the data

```python
x_train, x_test, y_train, y_test = train_test_split(df.text, df.status, random_state = 42, stratify = df.status)
```

## Word Embeddings

```python
total_words = len(set(corpus))
```

```python
tokenizer = Tokenizer(num_words = total_words, oov_token = "<OOV>")
tokenizer.fit_on_texts(x_train)
train_seq = tokenizer.texts_to_sequences(x_train)
test_seq = tokenizer.texts_to_sequences(x_test)
```

```python
padded_train = pad_sequences(train_seq, maxlen = 50, padding = 'post', truncating = 'post')
padded_test = pad_sequences(test_seq, maxlen = 50, truncating = 'post')
```

## Model

```python
model = Sequential()

model.add(Embedding(total_words, output_dim = 128))
model.add(Bidirectional(LSTM(64, return_sequences = True)))
model.add(Bidirectional(LSTM(32)))

model.add(Dense(128, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model.summary()
```

## Training the model

```python
tf.config.run_functions_eagerly(True)
```

```python
model.fit(padded_train, y_train,batch_size = 64, validation_split = 0.1, epochs = 10)
```

## Prediction

Since we used sigmoid function, everything above 0.5 will be considered as True or correct, and everything below 0.5 will be considered False or incorrect.

```python
pred = model.predict(padded_test)

prediction = []

for i in range(len(pred)):
    if pred[i].item() > 0.5:
        prediction.append(1)
    else:
        prediction.append(0)
```

## Accuracy

```python
accuracy = accuracy_score(list(y_test), prediction)
print("Model Accuracy : ", accuracy)
```

Very good accuracy. We should have used callbacks while training to avoid overfitting but thankfully that did not happen. I would not want to change anything about this model, but one interesting way to solve this problem can be to use pre-trained embeddings like GloVe instead of training everything from scratch, and then tune it to our dataset.
