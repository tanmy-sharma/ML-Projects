```python
from transformers import pipeline, set_seed

import matplotlib.pyplot as plt

import pandas as pd
from datasets import load_dataset, load_metric
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

import pandas as pd
import numpy as np

import nltk
from nltk.tokenize import sent_tokenize

nltk.download("punkt")

# importing everything required
```

Dataset used : CNN/Daily Mail Dataset

Important part of this dataset is that we get abstactive summararies, not extractive summaries.

```python
from datasets import load_dataset

dataset = load_dataset("cnn_dailymail", version="3.0.0")

print(f"Features in cnn_dailymail : {dataset['train'].column_names}")
```

```python
# trying out the sample of the data
sample = dataset["train"][1]
print(f"""
Article (excerpt of 500 characters, total length: {len(sample["article"])}):
""")
print(sample["article"][:500])
print(f'\nSummary (length: {len(sample["highlights"])}):')
print(sample["highlights"])
```

Text Summarization Pipelines

```python
sample_text = dataset["train"][1]["article"][:1000]
# we save sumamry in dictionary
summaries = {}
```

Summarization Baseline Model for comparision later on

```python
def baseline_summary_three_sent(text):
    return "\n".join(sent_tokenize(text)[:3])
```

```python
summaries['baseline'] = baseline_summary_three_sent(sample_text)

summaries['baseline']
```

# GPT - 2

1. GPT2 uses Byte Pair Encoding to create the tokens in its vocabulary. This means the tokens are usually parts of words.

2. GPT-2 was trained with the goal of causal language modeling (CLM) and is thus capable of predicting the next token in a sequence. GPT-2 may create syntactically coherent text by utilizing this capability.

3. GPT-2 generates synthetic text samples in response to the model being primed with an arbitrary input. The model is chameleon-like — it adapts to the style and content of the conditioning text.

4. The TL;DR symbol, which stands for “too lengthy; didn’t read,” is an ideal job designator for summarizing. The TL;DR symbol can be used as a padding element to shorten the review text and indicate to the model that the important content ends there.

```python
from transformers import pipeline, set_seed

set_seed(42)

pipe = pipeline('text-generation', model = 'gpt2-medium' )

gpt2_query = sample_text + "\nTL;DR:\n"

pipe_out = pipe(gpt2_query, max_length = 512, clean_up_tokenization_spaces = True)
```

```python
pipe_out
```

```python
pipe_out[0]["generated_text"][len(gpt2_query) :] #print summary
```

```python
summaries['gpt2'] = "\n".join(sent_tokenize(pipe_out[0]["generated_text"][len(gpt2_query) :]))
```

# T5

T5 (Text-To-Text Transfer Transformer) is a transformer model that is trained in an end-to-end manner with text as input and modified text as output, in contrast to BERT-style models that can only output either a class label or a span of the input. This text-to-text formatting makes the T5 model fit for multiple NLP tasks like Summarization, Question-Answering, Machine Translation, and Classification problems.

```python
pipe = pipeline('summarization', model = 't5-small' )

pipe_out = pipe(sample_text)
```

```python
pipe_out
```

```python
summaries['t5'] = 'n'.join(sent_tokenize(pipe_out[0]['summary_text']))
```

# BART

BART is a denoising autoencoder for pretraining sequence-to-sequence models. It is trained by following
1. Corrupting text with an arbitrary noising function
2. Learning a model to reconstruct the original text. It uses a standard Transformer-based neural machine translation architecture.

 BERT uses a standard seq2seq/NMT architecture with a bidirectional encoder (like BERT) and a left-to-right decoder (like GPT). This means the encoder's attention mask is fully visible, like BERT, and the decoder's attention mask is causal, like GPT2, which implies that a fine-tuned BART model can take a text sequence (for example, English) as input and produce a different text sequence at the output (for example, French).



```python
pipe = pipeline("summarization", model="facebook/bart-large-cnn")
pipe_out = pipe(sample_text)
```

```python
pipe_out
```

```python
summaries["bart"] = "\n".join(sent_tokenize(pipe_out[0]["summary_text"]))
```

# Testing Our Summaries using ROUGE

Rouge measures recall: how much the words (and/or n-grams) in the human reference summaries appeared in the machine generated summaries, for example ROUGE-n recall=40% means that 40% of the n-grams in the reference summary are also present in the generated summary.

```python
rouge_metric = load_metric('rouge')
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]

reference = dataset['train'][1]['highlights']

records = []

for model_name in summaries:
    rouge_metric.add(prediction = summaries[model_name], reference = reference )
    score = rouge_metric.compute()
    rouge_dict = dict((rn, score[rn].mid.fmeasure ) for rn in rouge_names )
    print('rouge_dict ', rouge_dict )
    records.append(rouge_dict)

pd.DataFrame.from_records(records, index = summaries.keys() )
```
